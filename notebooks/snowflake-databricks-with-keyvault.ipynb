```
# Use secrets DBUtil to get Snowflake credentials.
user = dbutils.secrets.get("jyotsnar-keyvault-scope", "snowflake-username")
password = dbutils.secrets.get("jyotsnar-keyvault-scope", "snowflake-password")

# snowflake connection options
options = {
  "sfUrl": "https://<test>.west-us-2.azure.snowflakecomputing.com",
  "sfUser": user,
  "sfPassword": password,
  "sfDatabase": "DEMO_DB",
  "sfSchema": "PUBLIC",
  "sfWarehouse": "COMPUTE_WH"
}
```

```
df = spark.read \
  .format("snowflake") \
  .options(**options) \
  .option("query",  "select * from demo_datamart") \
  .load()

df.show()
```
